# Dating data set

#### Display it

```python
properties, labels = kNN.load_dating_data_set()

plt.figure().set_size_inches(18, 6)
plt.subplot(1, 3, 1)
plot(properties, labels, 0, 1)
plt.xlabel("property 0");plt.ylabel("property 1")

plt.subplot(1, 3, 2)
plot(properties, labels, 0, 2)
plt.xlabel("property 1");plt.ylabel("property 2")

plt.subplot(1, 3, 3)
plot(properties, labels, 1, 2)
plt.xlabel("property 1");plt.ylabel("property 2")
plt.tight_layout()
```

#### After normalization

```python
properties, labels = kNN.load_dating_data_set()
properties = kNN.normalize_dataset(properties)

plt.figure().set_size_inches(18, 6)
plt.subplot(1, 3, 1)
plot(properties, labels, 0, 1)
plt.xlabel("property 0");plt.ylabel("property 1")

plt.subplot(1, 3, 2)
plot(properties, labels, 0, 2)
plt.xlabel("property 1");plt.ylabel("property 2")

plt.subplot(1, 3, 3)
plot(properties, labels, 1, 2)
plt.xlabel("property 1");plt.ylabel("property 2")
plt.tight_layout()
```

#### Testing the algorithm

```python
kNN.test_dating(3)
```

#### Error as a function of the fraction of test data

```python
ratios = np.linspace(0.1, 1, 9, endpoint=False)
errors = [kNN.test_dating(3, r) for r in ratios]
plt.plot(ratios, errors)
```

#### Error as a function of the fraction of the number of neighbors chosen to vote

```python
kneigh = np.arange(1, 51)
errors = [kNN.test_dating(k, 0.1) for k in kneigh]
plt.plot(kneigh, errors)
```
